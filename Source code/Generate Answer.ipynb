```python

# Tiền xử lý dữ liệu
def preprocess(example):
  # Nối câu hoi và nội dung thành một đầu vào duy nhất
  input_text = example['question'] + ' ' + example['content']
  # Định dạng đầu vào theo định dạng T5
  input_text = f'{input_text} </s>'
  # Mã hóa đầu vào
  input_encoding = tokenizer(input_text, max_length=1024, truncation=True,padding=True,return_tensors="pt")
  # Định dạng câu tra loi làm đầu ra mong muốn
  target_text = example['answer']
  # Mã hóa đầu ra mong muốn
  target_encoding = tokenizer([i for i in target_text], max_length=724, truncation=True,padding=True,return_tensors="pt")
  # Trả về đầu vào và đầu ra đã được mã hóa
  return {'input_ids': input_encoding['input_ids'], 'attention_mask': input_encoding['attention_mask'],
          'labels': target_encoding['input_ids']}

# Tiền xử lý tập huấn luyện và tập kiểm tra
train_dataset = df.apply(preprocess, axis=1)
#test_dataset = dataset['test'].map(preprocess)
```

```python
from datasets import load_metric
rouge = load_metric("rouge")

def compute_metrics(pred):
    labels_ids = pred.label_ids
    pred_ids = pred.predictions

    # all unnecessary tokens are removed
    pred_str = tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    labels_ids[labels_ids == -100] = tokenizer.pad_token_id
    label_str = tokenizer.batch_decode(labels_ids, skip_special_tokens=True)

    rouge_output = rouge.compute(predictions=pred_str, references=label_str, rouge_types=["rouge4"])["rouge4"].mid

    return {
        "rouge4_precision": round(rouge_output.precision, 4),
        "rouge4_recall": round(rouge_output.recall, 4),
        "rouge4_fmeasure": round(rouge_output.fmeasure, 4),
    }
```

#Fine-tunning

```python

# Fine-tune the model on ViQuAD v2
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model_1.to(device)
model_1.train()
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(train_dataset, [train_size, val_size])

optimizer = torch.optim.AdamW(model_1.parameters(), lr=1e-4)
num_epochs = 3

for epoch in range(num_epochs):
    for batch in train_dataset:
        # Move data to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        # Compute loss
        outputs = model_1(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss

        # Backpropagate loss and update weights
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

    # Evaluate on validation set
    model_1.eval()
    total_loss = 0

    for batch in val_dataset:
        with torch.no_grad():
            # Move data to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            # Compute loss
            outputs = model_1(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss

            total_loss += loss.item()

    avg_val_loss = total_loss / len(val_dataset)
    print(f'Epoch: {epoch + 1}, Validation Loss: {avg_val_loss:.4f}')
```
