```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq, Seq2SeqTrainer, TrainingArguments, Seq2SeqTrainingArguments
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained("/content/drive/MyDrive/DS204/model_Vit5_Answer")
```

```python
tokenizer = AutoTokenizer.from_pretrained("VietAI/vit5-base")
```

```python
df_a = pd.read_csv("/content/drive/MyDrive/DS204/Q&A/caption and answer_train.csv")
```

```python
df_q = pd.read_csv("/content/drive/MyDrive/DS204/Q&A/question_train_5.csv")
```

```python
a = []
for id, i in enumerate(df_a['captioan']):
  input_sentence = df_q['question'][id] + " " + i
  input_tokenized = tokenizer.encode(input_sentence, return_tensors="pt")
  output_tokenized = model.generate(input_tokenized)
  output_sentence = tokenizer.decode(output_tokenized[0], skip_special_tokens=True)
  print(output_sentence)
  a.append(output_sentence)
```
